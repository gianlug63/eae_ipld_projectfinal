```python
# The library you have to use
import numpy as np

# Some extra libraries to deal with images and files
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img_path = "../data/starry_night.png"

with Image.open(img_path) as img:
    img_arr = np.array(img)

plt.imshow(img_arr)
```

```python
# Ex 1.1: Get the height and width of the image in pixels

def get_max_dimensions(img_path):
    with Image.open(img_path) as img:
        width, height = img.size
        return width, height
img_path = "../data/starry_night.png"
max_height = get_max_dimensions(img_path)
max_width = get_max_dimensions(img_path)
print(f"Height: {max_height} pixels, height is the value on the right")
print(f"Width: {max_width} pixels,width is the value on the left")
```

```python
# Ex 1.2: Define the dimensions of the crop, make sure that min is smaller than max and also that min is larger than 0 
# and max is smaller than the size of the image for every dimension. If not, print a message explaining the error.

# Example crop dimensions, modify them to test your code
crop_min_h = 600 
crop_max_h = 1500
crop_min_w = 250
crop_max_w = 1100


# Constraint checks
image_width = 1377
image_height = 1089
error_messg = []
if crop_min_h >= crop_max_h:
    error_messg.append("Error: crop_min_h should be smaller than crop_max_h.")

if crop_min_w >= crop_max_w:
    error_messg.append("Error: crop_min_w should be smaller than crop_max_w.")

if crop_min_h <= 0 or crop_min_h > image_height:
    error_messg.append("Error: crop_min_h should be larger than 0 and not exceed the image height.")

if crop_max_h <= 0 or crop_max_h > image_height:
    error_messg.append("Error: crop_max_h should be larger than 0 and not exceed the image height.")

if crop_min_w <= 0 or crop_min_w > image_width:
    error_messg.append("Error: crop_min_w should be larger than 0 and not exceed the image width.")

if crop_max_w <= 0 or crop_max_w > image_width:
    error_messg.append("Error: crop_max_w should be larger than 0 and not exceed the image width.")

if error_messg:
    for error in error_messg:
        print(error)
else:
    print("Crop dimensions perfect.")
```

```python
# Ex 1.3: Generate the crop array into a new variable, use NumPy array slicing

crop_min_h = 600
crop_max_h = 1000
crop_min_w = 550
crop_max_w = 1100
img_array = np.random.random((1089, 1377, 4))
crop_arr =img_arr[crop_min_h:crop_max_h, crop_min_w:crop_max_w, :]
print(crop_arr) # TODO
```

```python
# Display the cropped image

plt.imshow(crop_arr)
```

```python
# Ex 1.4: Saving the cropped image into a new file, set the filename of the new image

crop_img = Image.fromarray(crop_arr)

file_name = 'image_cropped_sol'  # TODO: Set a file name, use a string

crop_img.save(f"../data/{file_name}.png")
```
